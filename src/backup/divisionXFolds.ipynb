```python
import numpy as np
import joblib
import pandas as pd
```

```python
# Importar los datos
label = joblib.load("/content/drive/My Drive/repo_tesis/data/label_W200_I50.joblib")

%cd "/content/drive/My Drive/repo_tesis/archivos_generados_codigos"
# Importo sorteo
sorteo = np.array(joblib.load('sorteo_sujetos.pkl'))

# Sé que los sujetos test son el 4 y el 7, los elegí arbitrariamente
# Los primeros dos sujetos del sorteo son los de validacion
```

```python
print(sorteo)
```

```python
# Cada elemento de la lista, sera una lista con los indices de los datos perteneciente a un sujeto
folds_ind = []
# Voy a recorrer cada sujeto en sorteo, e ir agregando al fold los indices de dato de ese sujeto
for i in sorteo:
  folds_ind.append(np.array(np.where(label[:, 2]==i)[0]))

joblib.dump(folds_ind, 'folds_W200_I50_sujeto.pkl')
```
