```python
import numpy as np
import joblib
import matplotlib.pyplot as plt

# Importar los datos
features = joblib.load("/content/drive/My Drive/repo_tesis/data/FEATURES_W200_I50.joblib")
label = joblib.load("/content/drive/My Drive/repo_tesis/data/label_W200_I50.joblib")
```

Fijo dos sujetos para test al azar. El resto los sorteo y tomo los dos primeros para validación. Si quisiera hacer validación cruzada, simplemente junto train con val y pongo en cada fold los datos de cada sujeto.

```python
# Calculo la cantidad de sujetos
label = np.array(label)
cantSujetos = np.max(label[:, 2])

# Identifico a los sujetos que voy a usar para testear
sujeto_test1 = 4
sujeto_test2 = 7

# Hago un sorteo a ver que sujeto va para cada conjunto
sorteo = np.random.permutation(cantSujetos) + 1

# Elimino del sorteo a los sujetos test porque los elijo yo
sorteo_sin_test = np.delete(sorteo, np.where(sorteo == sujeto_test1))
sorteo_sin_test = np.delete(sorteo_sin_test, np.where(sorteo_sin_test == sujeto_test2))

# El primer sujeto del sorteo va a test, el segundo a val y el resto a train
indices_test = list(np.where(label[:, 2]==sujeto_test1)[0]) # list(np.where(label[:, 2]==sorteo[0])[0])
indices_test.extend(np.where(label[:, 2]==sujeto_test2)[0])

indices_val = list(np.where(label[:, 2]==sorteo_sin_test[0])[0]) # indices_val = list(np.where(label[:, 2]==sorteo[2])[0])
# indices_test.extend(list(np.where(label[:, 2]==sorteo[1])[0]))
indices_val.extend(list(np.where(label[:, 2]==sorteo_sin_test[1])[0]))

indices_train = []
for j in sorteo_sin_test[2:]:
    indices_train.extend(np.where(label[:, 2]==j)[0])

%cd "/content/drive/My Drive/repo_tesis/archivos_generados_codigos"

joblib.dump(sorteo_sin_test, 'sorteo_sujetos.pkl')
joblib.dump(indices_train, 'indices_train_W200_I50_sujeto.pkl')
joblib.dump(indices_val, 'indices_val_W200_I50_sujeto.pkl')
joblib.dump(indices_test, 'indices_test_W200_I50_sujeto.pkl')
```
