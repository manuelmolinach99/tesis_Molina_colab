Este código devuelve los índices de features (son los mismos que label), de cada fold, para poder armar los folds fácilmente en el código que lo precise. Se decide no guardar directamente los datos particionados para ahorrar memoria, ya que se evita guardar datos repetidos pero ordenados de distinta manera.

```python
import numpy as np
import joblib
import pandas as pd
```

```python
# Importar los datos
label = joblib.load("/content/drive/My Drive/repo_tesis/data/label_W200_I50.joblib")

%cd "/content/drive/My Drive/repo_tesis/src/ronda_experimentos_1/particiones_1subTest"
```

```python
# convención: primer fold es el de test
sujetos_list = list(np.arange(0,10))

for i in sujetos_list:

  # Tengo que hacer que el sujeto que mando para test quede primero
  sujetos_aux = sujetos_list.copy()
  if i!=0:
    sujetos_aux.remove(i)   # remuevo el sujeto de test para insertarlo al principio (convencion para identificar sujeto de test)
    sujetos_aux.insert(0, i)  # agrego en la posicion 0 del arreglo el sujeto de test

# Cada elemento de la lista, sera una lista con los indices de los datos perteneciente a un sujeto
  folds_ind = []
  # Voy a recorrer cada sujeto en sorteo, e ir agregando al fold los indices de dato de ese sujeto
  for j in sujetos_aux:
    folds_ind.append(np.array(np.where(label[:, 2]==(j+1))[0]))

  joblib.dump(folds_ind, 'folds_W200_I50_sujetoTest_' + str(i+1) + '.pkl')
  joblib.dump(sujetos_aux, 'arraySujetos_folds_W200_I50_sujetoTest_' + str(i+1) + '.pkl')
```
