```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

scaler = StandardScaler()
scaler.fit(features)
data_scaled = scaler.transform(features)

pca = PCA(n_components=0.95)
data_red = pca.fit_transform(data_scaled)

print('dim antes PCA: ' + str(data_scaled.shape[1]))
print('dim luego PCA: ' + str(data_red.shape[1]))
print(pca.explained_variance_ratio_)
print(data_red.shape)
```

```python
# Guardar nuevos datos
%cd "/content/drive/My Drive/repo_tesis/data"
data_red = pd.DataFrame(data_red)
joblib.dump(data_red, "../data/FEATURES_W200_I50_PCA95.joblib")
joblib.dump(label, "../data/label_W200_I50.joblib")
```

Visualización de características

```python
label = np.array(label)
vector_gestos = label[:, 1]
ind_gesto1 = np.where(vector_gestos==1)[0]
ind_gesto2 = np.where(vector_gestos==2)[0]
ind_gesto3 = np.where(vector_gestos==3)[0]
ind_gesto4 = np.where(vector_gestos==4)[0]
ind_gesto5 = np.where(vector_gestos==5)[0]
ind_gesto6 = np.where(vector_gestos==6)[0]
ind_gesto7 = np.where(vector_gestos==7)[0]
ind_gesto8 = np.where(vector_gestos==8)[0]
ind_gesto9 = np.where(vector_gestos==9)[0]
```

```python
num_feature1 = 0
num_feature2 = 1
num_feature3 =2

x1 = data_red[ind_gesto1, num_feature1]
y1 = data_red[ind_gesto1, num_feature2]
z1 = data_red[ind_gesto1, num_feature3]

x2 = data_red[ind_gesto2, num_feature1]
y2 = data_red[ind_gesto2, num_feature2]
z2 = data_red[ind_gesto2, num_feature3]

x3 = data_red[ind_gesto3, num_feature1]
y3 = data_red[ind_gesto3, num_feature2]
z3 = data_red[ind_gesto3, num_feature3]

x4 = data_red[ind_gesto4, num_feature1]
y4 = data_red[ind_gesto4, num_feature2]
z4 = data_red[ind_gesto4, num_feature3]

x5 = data_red[ind_gesto4, num_feature1]
y5 = data_red[ind_gesto4, num_feature2]
z5 = data_red[ind_gesto4, num_feature3]

x6 = data_red[ind_gesto5, num_feature1]
y6 = data_red[ind_gesto5, num_feature2]
z6 = data_red[ind_gesto5, num_feature3]

# Truncar los datos para que tengan la misma longitud
min_length = min(len(x1), len(x2), len(x3))
x1_truncated, y1_truncated, z1_truncated = x1[:min_length], y1[:min_length], z1[:min_length]
x2_truncated, y2_truncated, z2_truncated = x2[:min_length], y2[:min_length], z2[:min_length]
x3_truncated, y3_truncated, z3_truncated = x3[:min_length], y3[:min_length], z3[:min_length]
x4_truncated, y4_truncated, z4_truncated = x4[:min_length], y4[:min_length], z4[:min_length]
x5_truncated, y5_truncated, z5_truncated = x5[:min_length], y5[:min_length], z5[:min_length]
x6_truncated, y6_truncated, z6_truncated = x6[:min_length], y6[:min_length], z6[:min_length]

# Lista de nombres personalizados para los conjuntos de datos
data_names = {'red': 'Clase 1', 'green': 'Clase 2', 'blue': 'Clase 3','lime': 'Clase 4', 'black': 'Clase 5', 'grey': 'Clase 6'}

# Combinar datos en una lista de listas
combined_data = [[x1_truncated[i], y1_truncated[i], z1_truncated[i], 'red'] for i in range(min_length)]
combined_data.extend([[x2_truncated[i], y2_truncated[i], z2_truncated[i], 'green'] for i in range(min_length)])
combined_data.extend([[x3_truncated[i], y3_truncated[i], z3_truncated[i], 'blue'] for i in range(min_length)])
combined_data.extend([[x4_truncated[i], y4_truncated[i], z4_truncated[i], 'lime'] for i in range(min_length)])
combined_data.extend([[x5_truncated[i], y5_truncated[i], z5_truncated[i], 'black'] for i in range(min_length)])
combined_data.extend([[x6_truncated[i], y6_truncated[i], z6_truncated[i], 'black'] for i in range(min_length)])



# Crear DataFrame
df = pd.DataFrame(combined_data, columns=['x', 'y', 'z', 'colors'])

# Crear el scatter plot 3D
fig = px.scatter_3d(df, x='x', y='y', z='z', color='colors',
                    title='Características en 3 dimensiones')

# Personalizar la leyenda
fig.for_each_trace(lambda t: t.update(name=data_names[t.name]))

# Mostrar el scatter plot
fig.show()
```
